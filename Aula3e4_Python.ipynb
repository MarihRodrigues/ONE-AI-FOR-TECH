```python
import google.generativeai as genai

def resumidor_de_emails(email_bodies):
  lista_de_resumos = []
  for numero, email in enumerate(email_bodies):
       # Usando o objeto gemini_model previamente inicializado
       resposta = gemini_model.generate_content(
           contents=f"""Vou te mandar o corpo de um
            e-mail. Quero que voce o resuma em apenas
             1 linha, passando o intuito daquele e-mail.
             Segue o e-mail: {email['corpo']}""" # Acessando o campo 'corpo' do dicionário
       )
       print(f"E-mail {numero + 1}: {resposta.text}")
       lista_de_resumos.append(f"E-mail {numero + 1}: {resposta.text}")
       print("-"*50)

  return lista_de_resumos
```

```python
lista_resumos = resumidor_de_emails(email_bodies)
```

Desafio:

1. Criar um arquivo .txt a partir de uma lista de perguntas vindas de uma lista em Python.
2. Ler as perguntas desse arquivo .txt e salva las em python
3. obter resposta de um LLM para cada uma
4. Salvar os resultados em um novo arquivo .csv
5. Ler os arquivos .csv usando Pandas

```python
#Etapa 1
lista_de_perguntas = [
    "De que e feito e sol",
    "De que e feito o planeta saturno",
    "De que e feito o planeta terra",
    "De que e feito o planeta marte",
    "De que e feito o planeta venus",
    "De que e feito o planeta mercurio",
    "De que e feito o planeta urano",
    "Qual a galaxia mais aniga ja encontrada",
    "Qual a maior estrela ja encontrada",
    "Qual a estrela mais proxima do sol"
]

with open("perguntas.txt", "w", encoding="utf-8") as arquivos:
  for pergunta in lista_de_perguntas:
    arquivos.write(pergunta + "\n")
```

```python
# Etapa 2

lista_desafio = []
with open("perguntas.txt", "r", encoding="utf-8" ) as arquivo:
     for linha in arquivo:
      lista_desafio.append(linha.strip())

      print(lista_desafio)
```

```python
lista_de_dicionarios_de_respostas = []

for pergunta in lista_desafio:
  resposta = gemini_model.generate_content(
      contents=f"Gere uma resposta muito sucinta para a pergunta {pergunta}"
  )

  lista_de_dicionarios_de_respostas.append({"pergunta": pergunta, "resposta": resposta.text})
```

```python
lista_de_dicionarios_de_respostas
```

```python
with open("respostas1.csv", "w", encoding="utf-8") as arquivo:
     arquivo.write("pergunta,resposta\n")
     for pergunta_dict in lista_de_dicionarios_de_respostas:
      arquivo.write(f"{pergunta_dict['pergunta']}, {pergunta_dict['resposta']}\n")
```

```python
df_perguntas_e_respostas = pd.DataFrame(lista_de_dicionarios_de_respostas)
```

```python
df_perguntas_e_respostas.to_csv("resposta2.csv", index=False, encoding="utf-8")
```
